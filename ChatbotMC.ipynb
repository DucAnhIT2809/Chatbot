```python
from tensorflow.tools import docs
import gensim
from gensim.parsing.preprocessing import strip_non_alphanum, strip_multiple_whitespaces, preprocess_string, split_alphanum, strip_short, strip_numeric
import re 
def lower_case(docs):
  return [doc.lower() for doc in docs]
def remove_punc(docs):
  return [strip_non_alphanum(doc).strip() for doc in docs]
docs = lower_case(df['Questions'])
docs = remove_punc(docs)
df['Question2'] = docs
print(docs)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer
tf=TfidfVectorizer()
#transformed train reviews
tf_train=tf.fit_transform(df["Question2"])
df_check=pd.DataFrame(tf_train.toarray(),columns=tf.get_feature_names())
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['Answers_code'] = le.fit_transform(df['Answers'])
```

```python
from sklearn.naive_bayes import MultinomialNB
mn=MultinomialNB()
mn.fit(df_check,df['Answers_code'])
```
